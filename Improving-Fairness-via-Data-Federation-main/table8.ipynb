```python
def dataSplit(train_data, test_data, client_split = ((.5, .2), (.3, .4), (.2, .4))):
    z1_idx = train_data[train_data.z == 1].index
    z0_idx = train_data[train_data.z == 0].index
    
    clients_idx = []
    z0l, z1l = 0, 0
    n0, n1 = len(z0_idx), len(z1_idx)
    for i in range(len(client_split)):
        if i < len(client_split) - 1:
            z0r, z1r = z0l + int(client_split[i][0]*n0), z1l + int(client_split[i][1]*n1)
        else:
            z0r, z1r = n0,  n1
        clients_idx.append( np.concatenate((z0_idx[z0l:z0r], z1_idx[z1l:z1r])) )
        random.shuffle(clients_idx[i])
        
    train_dataset = LoadData(train_data, "y", "z")
    test_dataset = LoadData(test_data, "y", "z")

    synthetic_info = [train_dataset, test_dataset, clients_idx]
    return synthetic_info
```

```python
def dataGenerate(seed = 432, train_samples = 3000, test_samples = 500, 
                y_mean = 0.6, client_split = ((.5, .2), (.3, .4), (.2, .4))):
    """
    Generate dataset consisting of two sensitive groups.
    """
    np.random.seed(seed)
    random.seed(seed)
        
    train_data, test_data = dataSample(train_samples, test_samples, y_mean, 2)
    return dataSplit(train_data, test_data, client_split)
```

```python
info1 = dataGenerate(seed = 123, test_samples = 2000, train_samples = 4667, client_split = ((.2, .2), (.2, .2), (.2, .2), (.4, .4)))

info2 = dataGenerate(seed = 123, test_samples = 1000, train_samples = 2333, client_split = ((.6, .2), (.4, .8)))
```
